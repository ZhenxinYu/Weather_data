# FCC API

```python
fips, county_name = [], []
```

```python
#use fcc api to get the fips code and county name
for i in tqdm(range(0, station.shape[0])):
    if i < len(fips) and i < len(county_name):
        continue
    lat = station.iloc[i,1]
    long = station.iloc[i,2]
    url = "https://geo.fcc.gov/api/census/block/find?latitude=" + str(lat) + "&longitude=" + str(long) + "&format=json"
    response = requests.get(url)
    try:
        response = eval(response.content.decode('utf-8'))["County"] 
    except:
        print((i, url))
        print(response.content.decode('utf-8'))
        fips.append(None)
        county_name.append(None)
        continue
    fips.append(response['FIPS'])
    county_name.append(response['name'])
```

```python
station['FIPS'] = pd.Series(fips)
station['County Name'] = pd.Series(county_name)
```

```python
#save the file in case the kernel dies
station.to_csv("station_fcc.csv", index = False)
```

```python
zipcode, county = [None]*len(station), [None]*len(station)
```

```python
for i in tqdm(range(0, station.shape[0])):
    if not pd.isnull(zipcode[i]) and not pd.isnull(county[i]):
        continue
    lat = float(station.iloc[i,1])
    long = float(station.iloc[i,2])
    #print((lat, long))
    try:
        result = gmaps.reverse_geocode((lat, long))
        #print(result)
        result_df = json_normalize(result)
        location = result_df.address_components[0]
        county_name = [l['long_name'][:-7] for l in location if l['types'][0] == 'administrative_area_level_2']
        zip_code = [l['long_name'] for l in location if l['types'][0] == 'postal_code']
    except:
        t = 0.1
        zip_code, county_name = [None], [None]
        while t<12:
            print("Sleep Time: ", t,"\tIndex: ", str(i))
            time.sleep(t)
            try:
                result = gmaps.reverse_geocode((lat, long))
                result_df = json_normalize(result)
                lcation = result_df.address_components[0]
                #county_name = [l['long_name'][:-7] for l in location if l['types'][0] == 'administrative_area_level_2']
                zip_code = [l['long_name'] for l in location if l['types'][0] == 'postal_code']
            except:
                t *= 2
            else:
                break
        zipcode[i] = zip_code[0]
        #county[i] = county_name[0]
    else:
        zipcode[i] = zip_code[0] if zip_code else None
        #county[i] = county_name[0] if county_name else None
```
