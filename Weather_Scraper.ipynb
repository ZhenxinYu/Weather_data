```python
import requests
import pandas as pd
import pickle
import os
```

# Query all the County

```python
county = []
offset, total = 0, 2000
while offset < total:
    token = 'GdNVaWoznDYioKmVaVCoCmolmXNvQqoI'
    url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/locations"
    url += '?limit=1000&offset={}'.format(offset)
    response = requests.get(url, headers={'token':token})

    query = eval(response.content.decode('utf-8'))
    total = query['metadata']['resultset']['count']
    for q in query['results']:
        if 'US' in q['name']:
            county.append(q)
    offset += 1000
    print(offset, total)
county = pd.DataFrame(county)
```

# Query all the stations

```python
stations = []
offset, total = 0, 2000
while offset < total:
    token = 'GdNVaWoznDYioKmVaVCoCmolmXNvQqoI'
    url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/stations"
    url += '?limit=1000&offset={}'.format(offset)
    response = requests.get(url, headers={'token':token})

    query = eval(response.content.decode('utf-8'))
    total = query['metadata']['resultset']['count']
    stations += query['results']
    offset += 1000
    print(offset, total)
stations = pd.DataFrame(stations)
```

# Store all the weather data
Iterate over all the desired date and store the data to corresponding subdirectory

```python
occ = set()
```

```python
for year in range(2006, 2018):
    for month in range(1, 13):
        for idx in range(len(county)):
            locationid, county_name = county['id'][idx], county['name'][idx]
            if not os.path.exists(county_name):
                os.mkdir(county_name)
            data = []
            offset, total = 0, 2000
            while offset < total:
                if (year, month, locationid) in occ:
                    break
                token = 'GdNVaWoznDYioKmVaVCoCmolmXNvQqoI'
                url = "https://www.ncdc.noaa.gov/cdo-web/api/v2/data"
                url += '?datasetid=GHCND&startdate={0:}-{1:02}-01&enddate={0:}-{1:02}-{2}'.format(year, month, 30 if month in [2,4,6,9,11] else 31)
                url += '&locationid={}&limit=1000&offset={}'.format(locationid, offset)
                response = requests.get(url, headers={'token':token})

                query = eval(response.content.decode('utf-8'))
                if not query:
                    break
                total = query['metadata']['resultset']['count']
                data += query['results']

                offset += 1000
            pd.DataFrame(data).to_csv('{}/{}-{}.csv'.format(county_name, year, month))
            occ.add((year, month, locationid))
```
